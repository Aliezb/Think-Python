```python
#增量开发实例
import math
def hypotenuse(a,b):
    a_2 = a**2
    b_2 = b**2
    c_2 = a_2 + b_2
    c = math.sqrt(c_2)
    return c
hypotenuse(3,4)
```

```python
def is_between(x,y,z):
    if x<=y and y<=z :
        return True
    else :
        return False
is_between(1,3,2)
```

```python
def ack(m,n):
    if m==0:
        s=n+1
    elif m>0 and n==0:
        s=ack(m-1,1)
    elif m>0 and n>0:
        s=ack(m-1,ack(m,n-1))
    return s
ack(3,4)
```
